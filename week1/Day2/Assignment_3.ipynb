```python
%sql
drop table if exists Employee;
drop table if exists Sales;
```

```python
%sql
CREATE TABLE if not exists Employee ( 
emp_id INT PRIMARY KEY, 
emp_name VARCHAR(50), 
department VARCHAR(50), 
salary DECIMAL(10, 2), 
joining_date DATE 
); 

CREATE TABLE if not exists Sales ( 
sales_id INT PRIMARY KEY, 
emp_id INT, 
product VARCHAR(50), 
amount DECIMAL(10, 2), 
sale_date DATE 
); 
```

## Find the total salary for each department in the Employee table. 


```python
%sql
SELECT department, SUM(salary)
FROM Employee
GROUP BY department;
```

## Count the number of employees in each department

```python
%sql
SELECT department,count(*) FROM Employee GROUP By department;
```

## Calculate the average salary of employees in each department. 

```python
%sql
Select department,avg(salary) FROM Employee GROUP BY department;
```

## Find the maximum salary in each department

```python
%sql
SELECT department,max(salary) FROM Employee GROUP BY department;
```

## Find the minimum salary in each department. 

```python
%sql
SELECT department,min(salary) FROM Employee GROUP BY department;
```

##  Find the total salary for departments where the total salary exceeds 100,000.

```python
%sql
SELECT department, SUM(salary)
FROM Employee
GROUP BY department
HAVING SUM(salary)>100000;
```

## Count the number of employees in departments with more than 2 employees

```python
%sql
SELECT department,count(*)  FROM Employee GROUP By department HAVING count(*)>2;
```

##Find the total sales amount for each employee.

```python
%sql
SELECT e.emp_name,SUM(S.amount) as total_sales_amount
FROM Employee e 
JOIN Sales S 
ON e.emp_id=S.emp_id 
GROUP BY e.emp_name;
```

## List the number of sales made by each employee

```python
%sql
SELECT e.emp_name,count(*) as number_Of_sales
FROM Employee e 
JOIN Sales S
on e.emp_id=S.emp_id
GROUP BY e.emp_name;
```

## Find the total sales amount grouped by product

```python
%sql
SELECT S.product,SUM(S.amount) FROM Sales S GROUP BY S.product;
```

## Calculate the average sales amount grouped by product.

```python
%sql
SELECT S.product,AVG(S.amount) FROM Sales S GROUP BY S.product;
```

## Find employees who have made more than 2 sales, grouped by their names.

```python
%sql
SELECT e.emp_name,count(*) as number_Of_sales
FROM Employee e 
JOIN Sales S
on e.emp_id=S.emp_id
GROUP BY e.emp_name
HAVING number_Of_sales>2;
```

## Calculate the total salary and total sales amount for each employee


```python
%sql
SElECT e.emp_name,SUM(e.salary),SUM(S.amount) as total_sales_amount
FROM Employee e 
JOIN Sales S 
ON e.emp_id=S.emp_id 
GROUP BY e.emp_name;
```

## Count the number of unique products sold by each employee. 


```python
%sql
SELECT e.emp_name, COUNT(DISTINCT s.product) AS unique_products
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

##  Find the highest sales amount made by each employee

```python
%sql
SELECT e.emp_name, MAX(s.amount) AS max_sale
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name;
```

## Calculate the total sales amount grouped by product and filtered by products where the total exceeds 50,000

```python
%sql
SELECT product, SUM(amount) AS total_sales
FROM Sales
GROUP BY product
HAVING SUM(amount) > 50000;
```

## Find the departments with the highest average sales amount

```python
%sql
SELECT e.department
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY AVG(s.amount) DESC
LIMIT 1;
```

## Find the department with the highest total sales amount

```python
%sql
SELECT e.department
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY SUM(s.amount) DESC
LIMIT 1;
```

##  Show the top 3 employees with the highest total sales amount, grouped by employee names

```python
%sql
SELECT e.emp_name, SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_name
ORDER BY total_sales DESC
LIMIT 3;
```

##  Calculate the total number of employees and the average salary, grouped by the year of joining.

```python
%sql
SELECT YEAR(joining_date) AS join_year,
       COUNT(*) AS total_employees,
       AVG(salary) AS avg_salary
FROM Employee
GROUP BY YEAR(joining_date);
```

## Find the total sales amount for each department (using a join between Employee and Sales). 


```python
%sql
SELECT e.department, SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.department;
```

## Show employees who have not made any sales, grouped by their department 


```python
%sql
SELECT e.department, e.emp_name
FROM Employee e
LEFT JOIN Sales s ON e.emp_id = s.emp_id
WHERE s.emp_id IS NULL
GROUP BY e.department, e.emp_name;
```
