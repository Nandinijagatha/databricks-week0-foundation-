```python
%sql

CREATE OR REPLACE TABLE employees (
    emp_id INT,
    full_name STRING,
    email STRING,
    department STRING,
    city STRING,
    salary STRING,
    remarks STRING
);


INSERT INTO employees
SELECT * FROM VALUES
(1, 'Karthik Kondpak', 'karthik.k@gmail.com', 'Data Engineering', 'Hyderabad', '75000', ' Top performer '),
(2, 'Veena Reddy', 'veena_r@company.com', 'Analytics', 'Bangalore', '65000', 'Excellent communication'),
(3, 'Ravi kumar', 'ravi.kumar@org.in', 'Data Science', 'Chennai', '85000', 'Needs improvement'),
(4, 'Anil', 'anil@abc.com', 'DEVOPS', 'Pune', '70000', NULL),
(5, ' Suresh ', 'suresh@xyz.com', 'data engineering', ' hyderabad ', '60000', ' ')
AS v(emp_id, full_name, email, department, city, salary, remarks);
```

```python
%sql
-- 1. LENGTH / CHAR_LENGTH
SELECT full_name, LENGTH(full_name) AS len FROM employees;
SELECT full_name, CHAR_LENGTH(full_name) AS char_len FROM employees;
```

```python
%sql

-- 2. UPPER / LOWER
SELECT UPPER(department) AS dept_upper,
       LOWER(city) AS city_lower
FROM employees;
```

```python
%sql

-- 3. TRIM / LTRIM / RTRIM
SELECT 
TRIM(full_name) AS trim_name,
LTRIM(full_name) AS ltrim_name,
RTRIM(full_name) AS rtrim_name
FROM employees;
```

```python
%sql
-- 4. CONCAT
SELECT CONCAT(full_name, ' - ', department) AS emp_details
FROM employees;
```

```python
%sql

-- 5. CONCAT_WS
SELECT CONCAT_WS(' | ', emp_id, full_name, city) AS details
FROM employees;
```

```python
%sql

-- 6. SUBSTRING / SUBSTR
SELECT SUBSTRING(email, 1, 7) AS sub1 FROM employees;
SELECT SUBSTR(email, 1, 5) AS sub2 FROM employees;
```

```python
%sql

-- 7. LEFT / RIGHT
SELECT LEFT(full_name, 4) AS left_name,
       RIGHT(city, 3) AS right_city
FROM employees;
```

```python
%sql

-- 8. INSTR
SELECT email, INSTR(email, '@') AS position
FROM employees;
```

```python
%sql

-- 9. LOCATE
SELECT email, LOCATE('.', email) AS dot_position
FROM employees;
```

```python
%sql
--10. REPLACE
SELECT REPLACE(department, 'Data', 'Big Data') AS updated_dept
FROM employees;
```

```python
%sql

-- 11. REVERSE
SELECT full_name, REVERSE(full_name) AS reversed_name
FROM employees;
```

```python
%sql

-- 12. LPAD / RPAD
SELECT LPAD(emp_id, 5, '0') AS padded_id FROM employees;
SELECT RPAD(city, 15, '*') AS padded_city FROM employees;
```

```python
%sql

-- 13. TRIM + REPLACE
SELECT TRIM(REPLACE(city, ' ', '')) AS clean_city
FROM employees;
```

```python
%sql

-- 14. IFNULL (use COALESCE in Spark)
SELECT full_name, COALESCE(remarks, 'No remarks') AS remarks
FROM employees;
```

```python
%sql

-- 15. COALESCE
SELECT full_name, COALESCE(remarks, 'N/A') AS remarks
FROM employees;
```

```python
%sql

-- 16. FIND_IN_SET (Spark alternative using array)
SELECT array_position(split('Data,Analytics,AI', ','), 'Analytics') AS position;
```
