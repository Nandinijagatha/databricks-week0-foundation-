```python

%sql
CREATE TABLE IF NOT EXISTS orders (
    order_id INT PRIMARY KEY,
    customer_name VARCHAR(50),
    order_date DATE,
    order_time TIMESTAMP,
    delivery_date DATE,
    order_amount DECIMAL(10,2)
);
```

```python
%sql
CREATE OR REPLACE TABLE orders AS
SELECT * FROM VALUES
(1, 'Karthik', '2024-01-15', '2024-01-15 10:30:45', '2024-01-20', 2500.00),
(2, 'Veena', '2024-02-18', '2024-02-18 18:45:20', '2024-02-22', 3200.50),
(3, 'Ravi', '2024-03-02', '2024-03-02 09:15:10', '2024-03-08', 4100.75),
(4, 'Anil', '2024-03-09', '2024-03-09 14:05:55', '2024-03-15', 1800.00),
(5, 'Suresh', '2024-01-07', '2024-01-07 23:55:00', '2024-01-12', 2900.00)
AS t(order_id, customer_name, order_date, order_timestamp, delivery_date, order_amount);
```

## EXTRACTING YEAR, MONTH, DAY 

```python
%sql


SELECT 

YEAR(order_date), 

MONTH(order_date), 

DAY(order_date) 

FROM orders; 
```

## EXTRACT USING EXTRACT() 


```python
%sql
SELECT 

EXTRACT(YEAR FROM order_date), 

EXTRACT(MONTH FROM order_date), 

EXTRACT(DAY FROM order_date) 

FROM orders; 
```

## MONTH NAME AND DAY NAME 



```python
%sql
SELECT 

MONTHNAME(order_date), 

DAYNAME(order_date) 

FROM orders; 
```

##  WEEKDAY AND DAY OF WEEK 

```python
%sql
SELECT 

WEEKDAY(order_date), 

DAYOFWEEK(order_date) 

FROM orders; 
```

## Identifying weekends

```python
%sql
SELECT order_id, order_date
FROM orders
WHERE DAYNAME(order_date) IN ('Saturday', 'Sunday');
```

## Identifying Weekends using DAYOFWEEK()

```python
%sql
SELECT order_id, order_date,DAYOFWEEK(order_date)
FROM orders
WHERE DAYOFWEEK(order_date) IN (1, 7);
```

##  IDENTIFY WEEKDAYS 

```python
%sql
SELECT order_id, order_date 

FROM orders 

WHERE DAYOFWEEK(order_date) BETWEEN 2 AND 6; 
```

## DATE ARITHMETIC

```python
%sql
SELECT order_date, DATE_ADD(order_date, 5) AS new_date
FROM orders;
```

```python
%sql
SELECT order_date, DATE_SUB(order_date, 3 )

FROM orders; 
```

## DATEDIFF

```python
%sql
SELECT 

order_id, 

DATEDIFF(delivery_date, order_date) AS delivery_days 

FROM orders; 
```

## TIMESTAMPDIFF

```python
%sql
SELECT 

TIMESTAMPDIFF(DAY, order_date, delivery_date) AS days_diff, 

TIMESTAMPDIFF(MONTH, order_date, delivery_date) AS months_diff 

FROM orders; 
```

```python
%sql

SELECT LAST_DAY(order_date) AS last_day
FROM orders;
```

##  FIRST DAY OF MONTH

```python
%sql
SELECT DATE_SUB(order_date, INTERVAL DAY(order_date)-1 DAY) AS first_day
FROM orders;
```

## DATE FORMAT

```python
%sql
SELECT DATE_FORMAT(order_date, '%d-%m-%Y') AS format1
FROM orders;

SELECT DATE_FORMAT(order_date, '%M %d, %Y') AS format2
FROM orders;

```

## FILTER BY MONTH

```python
%sql
SELECT *
FROM orders
WHERE MONTH(order_date) = 1;

SELECT *
FROM orders
WHERE MONTHNAME(order_date) = 'February';

```

## FINANCIAL YEAR

```python
%sql

SELECT order_date,
       CASE
           WHEN MONTH(order_date) >= 4 
                THEN CONCAT(YEAR(order_date), '-', YEAR(order_date)+1)
           ELSE CONCAT(YEAR(order_date)-1, '-', YEAR(order_date))
       END AS financial_year
FROM orders;

```

## REAL-TIME USE CASES

### Last 7 days

```python
%sql

SELECT *
FROM orders
WHERE order_date >= CURDATE() - INTERVAL 7 DAY;
```

###  Today orders

```python
%sql
SELECT *
FROM orders
WHERE DATE(order_timestamp) = CURDATE();
```
