```python
%sql

CREATE OR REPLACE TABLE employee_salary (
    emp_id INT,
    emp_name STRING,
    base_salary DECIMAL(10,2),
    bonus DECIMAL(10,2),
    tax_percent DECIMAL(5,2),
    experience_years INT
);

INSERT INTO employee_salary
SELECT * FROM VALUES
(1, 'Karthik', 75000.75, 5000.50, 10.00, 6),
(2, 'Veena', 65000.40, 4000.25, 8.50, 4),
(3, 'Ravi', 85000.90, 6000.75, 12.00, 8),
(4, 'Anil', 70000.10, NULL, 9.00, 5),
(5, 'Suresh', 60000.55, 3000.30, 7.50, 3)
AS v(emp_id, emp_name, base_salary, bonus, tax_percent, experience_years);
```

```python
%sql

-- 1. ABS – Absolute value
SELECT ABS(-100);
```

```python
%sql

-- 2. ROUND – rounding
SELECT emp_name, ROUND(base_salary, 0) AS round_0,
       ROUND(base_salary, 2) AS round_2
FROM employee_salary;
```

```python
%sql

-- 3. CEIL – round up
SELECT emp_name, CEIL(base_salary) AS ceil_val
FROM employee_salary;
```

```python
%sql

-- 4. FLOOR – round down
SELECT emp_name, FLOOR(base_salary) AS floor_val
FROM employee_salary;
```

```python
%sql

-- 5. TRUNCATE – remove decimals (Databricks: use TRUNC)
SELECT emp_name, TRUNCATE(base_salary, 1) AS trunc_val
FROM employee_salary;
```

```python
%sql

-- 6. MOD – remainder
SELECT emp_name, MOD(experience_years, 2) AS remainder
FROM employee_salary;
```

```python
%sql

-- 7. POWER
SELECT POWER(2,3) AS power_val, POW(5,2) AS pow_val;
```

```python
%sql

-- 8. SQRT
SELECT SQRT(64) AS sqrt_val;
```

```python
%sql

-- 9. SIGN
SELECT emp_name, SIGN(base_salary) AS sign_val
FROM employee_salary;
```

```python
%sql
-- 10. RAND
SELECT RAND() AS random_val;
```

```python
%sql
-- 11. FORMAT (Databricks → use format_number)
SELECT emp_name, format_number(base_salary, 2) AS formatted_salary
FROM employee_salary;
```

```python
%sql
-- 12. GREATEST
SELECT emp_name,
GREATEST(base_salary, COALESCE(bonus, 0)) AS max_val
FROM employee_salary;
```

```python
%sql
-- 13. LEAST
SELECT emp_name,
LEAST(base_salary, COALESCE(bonus, 0)) AS min_val
FROM employee_salary;
```
